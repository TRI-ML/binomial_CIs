# **Visualization of CIs**

This notebook visualizes
1. the mixed monotonic form of expected shortage
2. expected shortage vs true value of $p$
3. the mixed monotonic form of expected width
4. expected width vs true value of $p$.

We do not give any visualizations for expected excess because all computations are essentially the same as for expected shortage (i.e. finding a lower bound on failure probability is equivalent to finding an upper bond of success probability).

Also, while expected shortage and expected width appear to be concave functions of $p$, it is difficult to prove this holds for all values of $\alpha$ and $n$. Thus, without a proof of concavity we use the mixed-monotonic property of these functions to globally maximize them with a branch and bound search.

```python
import numpy as np
import sys; sys.path.insert(0, '..')
from binomial_cis import plot_expected_shortage_mixed_monotonic, plot_shortage_curve
from binomial_cis import plot_expected_width_mixed_monotonic, plot_width_curve
```

#### Visualize Mixed Monotic form of Expected Shortage

```python
n = 20
alpha = 0.05
```

```python
p1s = np.linspace(0, 1, num=15)
p2s = p1s
plt_esmm = plot_expected_shortage_mixed_monotonic(alpha, n, p1s, p2s)
```

#### Visualize Expected Shortage vs $p$

```python
plt_sc = plot_shortage_curve(alpha, n, verbose=False, randomized=True)
```
